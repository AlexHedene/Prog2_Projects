## Fast-fission factor - $\varepsilon$
The fast fission factor is the ratio of the fast neutrons produced by fissions at all energies to the number of fast neutrons produced in thermal fission. From the picture we can see that we start with 1000 nuetrons, and have 1030 neutrons after adding the neutrons from fast fission. The fast-fission factor can then be calculated as 1030/1000 = 1.3. This factor should be placed in the circle with fast neutrons and fast fission should be equal to 1030-1000 = 30.

## Resonance escape probability - p
The resonance escape probability is the probability that a neutron will be slowed down to thermal energy and espace resonance absorption. This probability can be calculated as the fraction of neutrons that reach thermal energies devided by the number of fast neutrons. From the picture we can see that we have 1026 fast neutrons where 930 sruvives resonance absorptions. The resonance escape probability can then be calculated as $p = \frac{930}{1026} = 0.906$ where 96 neutrons were captured in the reconance region. 

## Thermal Utilization Factor - f
The thermal utilization factor, f, is the fraction of thermal neutrons that are absorbed in the nuclear fuel, across all its isotopes. It describes how effectively the fuel absorbs thermal neutrons for sustaining the chain reaction.
From the picture the thermal utilization factor can be calculated from the raio between the Thermalized neutrons and the neutrons absorbed in the fuel. This results in: $f = \frac{858}{930} = 0.92$, meaning that 72 neutrons are lost due to parasitic capture.

## Reproduction Factor - $\eta$
The reproduction factor $\eta$ is defined as the ratio of the number of fast neutrons produced by thermal fission to the number of thermal neutrons absorbed in the fuel. From the picture we can see that we have 858 absorbed neutrons in the fuel which results in 1000 thermal neutrons. This gives a reproduction factor: $\eta = \frac{1000}{858} = 1.66$
